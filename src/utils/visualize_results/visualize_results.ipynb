```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

base_path = "/home/itai/research/PersonalValuesGeometry/figures/embeddings/num_sumple_5000/llama-3.2-3b-instruct"
num_layer = 28
```

```python
def display_images(base_path: str, i_layer: int, norm_type: str,):
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    prompt_types = ["explicit", "theme", "bare"]

    for idx, prompt_type in enumerate(prompt_types):
        post_fix = f"layer{i_layer}/valuenet/pos2neg/{norm_type}/{prompt_type}"
        file_name = "concept_direction_LOO.png"
        file_path = f"{base_path}/{post_fix}/{file_name}"
        try:
            img = mpimg.imread(file_path)
        except:
            img = None
        axes[idx].imshow(img)
        axes[idx].set_title(f'Prompt Type: {prompt_type}')
        axes[idx].axis('off')

    plt.tight_layout()
    plt.show()
```

# Raw

```python
prompt_type = "explicit"
norm_type = "base"

print("Norm Type: ", norm_type)
for i_layer in range(20, num_layer+1):
    print(f"==== Layer: {i_layer} =====")
    post_fix = f"layer{i_layer}/valuenet/pos2neg/{prompt_type}/{norm_type}/"
    try:
        display_images(base_path=base_path, i_layer=i_layer, norm_type=norm_type)
    except TypeError as e:
        print(f"Error loading images for Layer {i_layer}: {e}")
```

```python
prompt_type = "explicit"
norm_type = "norm_sentence_structure"

print("Norm Type: ", norm_type)
for i_layer in range(num_layer):
    print(f"==== Layer: {i_layer} =====")
    post_fix = f"laeyr{i_layer}/valuenet/pos2neg/{prompt_type}/{norm_type}/"
    file_name = "concept_direction_LOO.png"
    file_path = f"{base_path}/{post_fix}/{file_name}"
    display_images(base_path=base_path, i_layer=i_layer, norm_type=norm_type)
```

# PCA

```python
def display_pca_images(base_path: str, i_layer: int, norm_type: str,):
    filename = "concept_direction_LOO.png"
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    prompt_types = ["explicit", "theme", "bare"]

    for idx, prompt_type in enumerate(prompt_types):
        post_fix = f"layer{i_layer}/valuenet/pos2neg/{norm_type}/{prompt_type}/ica"
        file_path = f"{base_path}/{post_fix}/{filename}"
        try:
            img = mpimg.imread(file_path)
        except:
            img = None
        axes[idx].imshow(img)
        axes[idx].set_title(f'Prompt Type: {prompt_type}')
        axes[idx].axis('off')

    plt.tight_layout()
    plt.show()
```

```python
norm_type = "base"

print("Norm Type: ", norm_type)
for i_layer in range(1, num_layer):
    print(f"==== Layer: {i_layer+1} =====")
    display_pca_images(base_path=base_path, i_layer=i_layer+1, norm_type=norm_type)
```

```python
norm_type = "norm_sentence_structure"

print("Norm Type: ", norm_type)
for i_layer in range(0, num_layer):
    print(f"==== Layer: {i_layer+1} =====")
    try:
        display_pca_images(base_path=base_path, i_layer=i_layer+1, norm_type=norm_type)
    except:
        print("file not found")
```
